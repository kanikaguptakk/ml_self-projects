## Getting our data ready to be used with ML
##### Three main things we have to do:
### 1. Split the dat into features and labels (usually 'X' nd 'Y')
### 2. Filling (also called imputing) or disregarding missing values
### 3. Converting non-numerical values to numerical values (also calles feature encoding)

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
car_sales=pd.read_csv("data/car-sales-extended.csv")
car_sales.head(5)
```

```python
len(car_sales)
```

```python
car_sales["Doors"].value_counts()        # Doors will treated as categories
```

```python
car_sales.dtypes
```

```python
# Split into X and Y
X=car_sales.drop("Price", axis=1)
Y=car_sales["Price"]

# Split into training and test
X_train, Y_train, X_test, Y_test= train_test_split(X,Y, test_size=0.2)
```

```python
# Build ml model
from sklearn.ensemble import RandomForestRegressor   
# we want to predict a number that'why using regressor 
model= RandomForestRegressor()     # create a model
model.fit(X_train, Y_train)
model.score(X_test, Y_test)

```

```python
# One Hot Encoding- A process used to turn categories into  numbers
# ex., car-colour            car-red-green-blue
    #   0-red                 0-1-0-0
```

```python
# Turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)], remainder="passthrough")
transformed_X=transformer.fit_transform(X)
transformed_X
```

```python
dummies=pd.get_dummies(car_sales[["Make", "Colour", "Doors"]])
dummies
```

```python
# Let's refit the model
np.random.seed(42)
X_train, X_test, Y_train, Y_test=train_test_split(transformed_X, Y, test_size=0.2)
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```
