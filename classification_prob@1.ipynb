#### 0. An end-to-end Scikit-Learn workflow  (above mentioned)

```python
# 1. Get the data ready
import pandas as pd
import numpy as np
heart_disease=pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# Create X (feature matrix)
X= heart_disease.drop("target", axis=1)        # except target column
# Create y (labels)
Y=heart_disease["target"]
```

```python
# 2. Choose the right estimator/algorithm for our problems
# Choose the right model and hyperparameters (tile on a model)
from sklearn.ensemble import RandomForestClassifier
clf= RandomForestClassifier(n_estimators=100)   # clf (classification)=model

# we'll keep the default hyperparameters
clf.get_params()
```

```python
# 3.Fit the model to the training data
from sklearn.model_selection import train_test_split    # split the data into train and test
X_train, X_test, Y_train, Y_test=train_test_split (X, Y, test_size=0.2)       # data for training is 80%
```

```python
clf.fit (X_train, Y_train);
```

```python
X_train
```

```python
X_test
```

```python
Y_preds=clf.predict(X_test)
Y_preds
```

```python
Y_test
```

```python
# make a prediction
Y_label= clf.predict(np.array([0,1,2,3])
```

```python
# 4. Evaluate the model on the training data and the test data
clf.score(X_train, Y_train)
```

```python
clf.score(X_test, Y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test, Y_preds))
```

```python
confusion_matrix(Y_test, Y_preds)
```

```python
accuracy_score(Y_test, Y_preds)
```

```python
# 5. Improve a model 
# Try different amounts of n_estimators
np. random.seed(43)
for i in range(10,100,10):
    print(f"Trying model with{i} estimators...")
    clf=RandomForestClassifier(n_estimators=i).fit(X_train, Y_train)
    # or 
    # clf.fit(X_train, Y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, Y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and laod it
import pickle
pickle.dump(clf, open("Random_forest_model_1.pkl","wb"))      # wb- write in binary
```

```python
loaded_model=pickle.load(open("Random_forest_model_1.pkl", "rb"))
# rb- read binary
loaded_model.score(X_test, Y_test)
```
