## Choosing the right estimators/algorithm for our problem
some things to note:
* ScikitLearn refers to ml models, algorithms as estimators
* Classification problem- predicting a category (heart disease or not)
         -Sometimes u'll see clf (short of classifier) used as a classification estimator
* Regression problem- predicting a number (sale price of a car)
         -Sometimes u'll see model/reg (short of regressive) used as a regressiom estimator
                  

If u r working on a ml problem n looking to use SkLearn n not sure what model u should use, refer to SkLearn ml maps (cheatsheet)-
https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html


### Picking a ml model for a regression problem
Let's use the California Housing Dataset- https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html#sklearn.datasets.fetch_california_housing

```python
# Get California Housing Dataset
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
housing
```

```python
import pandas as pd
housing_df=pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df
```

```python
housing_df["target"]=housing["target"]
housing_df
```

```python
# Import Algorithm/Estimator (for SkLearn)
# model 1 (hit and trial)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
# Setup Random Seed
import numpy as np
np.random.seed(42)

# Create the data
X= housing_df.drop("target", axis=1)
Y= housing_df["target"]

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2)

# Instantiate and fit the model (on the training dataset)
model=Ridge()
model.fit(X_train, Y_train)

# Check the score of the model on the test dataset
model.score(X_test, Y_test)
```

 SGDRegressor is well suited for regression problems with a large number of training samples (> 10.000), for other problems we recommend Ridge, Lasso, or ElasticNet.

```python
# model 2 (hit and trial)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
# Setup Random Seed
import numpy as np
np.random.seed(42)

# Create the data
X= housing_df.drop("target", axis=1)
Y= housing_df["target"]

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2)

# Instantiate and fit the model (on the training dataset)
model=SGDRegressor()
model.fit(X_train, Y_train)

# Check the score of the model on the test dataset
model.score(X_test, Y_test)
```

```python
# model 3 (hit and trial)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
# Setup Random Seed
import numpy as np
np.random.seed(42)

# Create the data
X= housing_df.drop("target", axis=1)
Y= housing_df["target"]

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2)

# Instantiate and fit the model (on the training dataset)
model=Lasso()
model.fit(X_train, Y_train)

# Check the score of the model on the test dataset
model.score(X_test, Y_test)
```

```python
# model 4 (hit and trial)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoLars
# Setup Random Seed
import numpy as np
np.random.seed(42)

# Create the data
X= housing_df.drop("target", axis=1)
Y= housing_df["target"]

# Split into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2)

# Instantiate and fit the model (on the training dataset)
model=LassoLars()
model.fit(X_train, Y_train)

# Check the score of the model on the test dataset
model.score(X_test, Y_test)
```

What if Ridge didn't work or the score didn't find our needs?
Well, we could always try a different model...
How about we try an ensemble model (an ensemble is a combination of smaller SkLearn models to make better predictions than just a single model)?
SkLearn's ensemble model can be found here- https://scikit-learn.org/stable/modules/ensemble.html

```python
# Import the RandomForestRegressor model class from the ensemble module
from sklearn.ensemble import RandomForestRegressor

# Setup Random Seed
np.random.seed(42)

# Create the data
X=housing_df.drop("target", axis=1)
Y=housing_df["target"]

# Split into train and test sets
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2)

# Create RandomForest Model
model= RandomForestRegressor()
model.fit(X_train, Y_train)

# Check the score of the model on  the test set
model.score(X_test, Y_test)
```
